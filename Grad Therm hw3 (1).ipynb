Want to create a random walk to answer question 3.36, to start use the simple random walk provided on blueline. Following my work on question 3.36, I also answer question 3.70 in this program

```python
import numpy as np
import matplotlib.pyplot as plt
N = 1000
d = 1
finalposition=[]
disp = [0] #start the walk at 0
# np.random.randint gives either 0 or 1,
# so (*2) and (-1) give either -1 or 1.
each_walk = np.random.randint(2,size=N)*2 - 1
disp.extend(np.cumsum(d*each_walk))
finalposition.append(disp[-1])
plt.plot(disp,color='r')
plt.ylabel('Displacement')
plt.xlabel('Time Step')
plt.title('Single Random Walk')
plt.show()
print(finalposition)
```

a. The results from a single random walk do not matter as they do not tell us any meaningful statsitics. <br>               b. Also even if we run over multiple trials we are not guarenteed to reproduce the exact result, just our probability of producing one goes up as we use more trials. <br> Now edit the code to run over 1000 trials and varying steps while displaying final results in a histogram.

```python
N_particles = 1000
N1 = 10 #duplicating the sample code for different number of steps
N2 = 100
N3 = 1000
d = 1
finalposition1=[]
finalposition2=[]
finalposition3=[]
startposition = [0] * N_particles #needed to start at 0 for my arrays
# np.random.randint gives either 0 or 1,
# so (*2) and (-1) give either -1 or 1.
each_walk1 = np.random.randint(2,size=(N1,N_particles))*2 - 1
each_walk2 = np.random.randint(2,size=(N2,N_particles))*2 - 1
each_walk3 = np.random.randint(2,size=(N3,N_particles))*2 - 1
disp1 = np.cumsum(d*each_walk1,axis=0)
disp2 = np.cumsum(d*each_walk2,axis=0)
disp3 = np.cumsum(d*each_walk3,axis=0)
disp1 = np.vstack([startposition,disp1]) #putting the start position 
disp2 = np.vstack([startposition,disp2])
disp3 = np.vstack([startposition,disp3])
finalposition1.append(disp1[-1,:]) #saving the final position of each run to a list
finalposition2.append(disp2[-1,:])
finalposition3.append(disp3[-1,:])
max_disp1 = np.abs(np.max(disp1))
max_disp2 = np.abs(np.max(disp2))
max_disp3 = np.abs(np.max(disp3))
plt.plot(disp1,color='blue')
plt.ylim(np.array([-1,1])*1.5*max_disp1)
plt.ylabel('Displacement')
plt.xlabel('Time Step')
plt.title('Multiple Random Walks (10 steps)')
plt.show()
plt.plot(disp2,color='blue')
plt.ylim(np.array([-1,1])*1.5*max_disp2)
plt.ylabel('Displacement')
plt.xlabel('Time Step')
plt.title('Multiple Random Walks (100 steps)')
plt.show()
plt.plot(disp3,color='blue')
plt.ylim(np.array([-1,1])*1.5*max_disp3)
plt.ylabel('Displacement')
plt.xlabel('Time Step')
plt.title('Multiple Random Walks (1000 steps)')
plt.show()

#making histograms
plt.hist(finalposition1, align = 'left')
plt.xlabel('Final Position')
plt.ylabel('Frequency')
plt.title('Frequency vs Position (10 steps)')
plt.show()
plt.hist(finalposition2, align = 'left')
plt.xlabel('Final Position')
plt.ylabel('Frequency')
plt.title('Frequency vs Position (100 steps)')
plt.show()
plt.hist(finalposition3, align = 'left')
plt.xlabel('Final Position')
plt.ylabel('Frequency')
plt.title('Frequency vs Position (1000 steps)')
plt.show()
```

Now begin on problem 3.70, regarding a drunkard taking steps to the right or left on multiple nights<br>
Part a:

```python
import math
rightsteps = [14,13,11,12,11,12,16,16,14,8]
stepstaken = 20

def square(list):
    return [i ** 2 for i in list]
rightstepssquared = square(rightsteps)

def Average(list): 
    return sum(list) / len(list) 
mean = Average(rightsteps)
meansquared = Average(rightstepssquared)
squareofmeans = mean**2
variance = math.sqrt(meansquared - squareofmeans)
print("The mean is", mean) 
print("The mean squared is", meansquared)
print("The square of the means is", squareofmeans)
print("The variance is", variance)


def prob(list):
    return sum(list) / (len(list) * stepstaken)
probright = prob(rightsteps)
print("The probability of a step to the right based on adding the total steps to the right and dividing by the total steps taken is")
print(probright)

#rearranging the relative width equation and substituting q = 1-p allows us to create an equation for p
#this is the second way to estimate p
prob2right = (math.sqrt(stepstaken)*variance / mean) ** 2 +1
prob2right = 1/prob2right
print("The probability using the relative width of a step to the right is")
print(prob2right)
```

The 2 probabilities are close, but there is a notable difference, the one that comes using the relative width is probably more accurate considering it takes into account how the variance corresponds with the averages.
